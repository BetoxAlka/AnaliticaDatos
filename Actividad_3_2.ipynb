```python
#Importamos librerías requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carga desde un archivo .xlsx sin indice
df = pd.read_excel('Datos de Facturación.xlsx')
df
```

```python
#Reemplazo de datos de tipo string a datos numericos
df.CVE_CLPV = df.CVE_CLPV.replace({'O':'0','6781|':'6781','3612C':'3612','      ':'0','2385A':'2385','2510A':'2510','8440Q':'8440','M':'0'}, regex=False)
```

```python
#Conversion de tipo de dato de una columna
df['CVE_CLPV'] = df['CVE_CLPV'].astype(int)
```

```python
filtro1 = df[(df['CVE_CLPV'] >= 1000) & (df['CVE_CLPV'] <= 2000)] #Muestra registros con CVE_CLPV entre 1000 y 2000
filtro1
```

```python
#Convertir archivo filtrado a csv
filtro1.to_csv('Filtro1.csv')

from google.colab import files

files.download('Filtro1.csv')
```

```python
filtro2 = df[~(df['CVE_VEND'] == 5.0) & ~(df['CVE_VEND'] == 4.0)] #Muestra registros cuyo CVE_VEND NO sea 5 o 4
filtro2
```

```python
#Convertir archivo filtrado a csv
filtro2.to_csv('Filtro2.csv')

from google.colab import files

files.download('Filtro2.csv')
```

```python
#Conversion de tipo de dato de una columna
df['FECHA_ENT'] = df['FECHA_ENT'].astype(str)
```

```python
filtro3 = df[df['FECHA_ENT'] == '2022-02-28 00:00:00'] #Muestra registros donde la FECHA_ENT sea 28/02/2022
filtro3
```

```python
#Convertir archivo filtrado a csv
filtro3.to_csv('Filtro3.csv')

from google.colab import files

files.download('Filtro3.csv')
```

```python
filtro4 = df[(df['CAN_TOT'] < 5951.7) | (df['STATUS'] == 'E')] #Muestra registros donde la CAN_TOT sea menor a 5951.7 o su STATUS sea E
filtro4
```

```python
#Convertir archivo filtrado a csv
filtro4.to_csv('Filtro4.csv')

from google.colab import files

files.download('Filtro4.csv')
```
