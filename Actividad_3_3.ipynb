```python
#Carga desde un archivo .xlsx sin indice
df = pd.read_excel('Inmuebles.xlsx')
df.head(5)
```

```python
#Eliminar primera fila o filas innecesarias
df = pd.read_excel('Inmuebles.xlsx',skiprows=1) #Al generar el frame, se ignora la fila con las columnas UNNAMED
df
```

```python
#Eliminar columnas innecesarias
#axis = 1 para columnas
#axis = 0 para filas
df1 = df.drop(['Referencia'], axis=1) #El nuevo frame NO contiene la columna 0, por lo que falta eliminar la columna Referncia
df1
```

```python
#Convertir archivo filtrado a csv
df1.to_csv('Filtro1.csv')

from google.colab import files

files.download('Filtro1.csv')
```

```python
filtro2 = df1[(df1['Fecha_Alta'] == '2007-04-29') | (df1['Fecha_Alta'] == '2004-01-02')] #Filtrado por las fechas 2007-04-29 y 2004-01-02
filtro2
```

```python
#Convertir archivo filtrado a csv
filtro2.to_csv('Filtro2.csv')

from google.colab import files

files.download('Filtro2.csv')
```

```python
filtro3 = df1[(df1['Tipo'] == 'Parking') | (df1['Tipo'] == 'Local') | (df1['Tipo'] == 'Industrial') | (df1['Tipo'] == 'Piso')]
filtro3 #Filtro mediante tipo de parking, local, industrial y piso
```

```python
#Convertir archivo filtrado a csv
filtro3.to_csv('Filtro3.csv')

from google.colab import files

files.download('Filtro3.csv')
```
