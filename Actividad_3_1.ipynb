```python
#Importamos librerías requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carga desde un archivo .xlsx sin indice
df = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df
```

```python
#Filtro por objeto
filtro1 = df[df['FECHA_DOC'] == '2022-01-05'] #Se filtran los registros en base a la fecha establecida
filtro1
```

```python
#Convertir archivo filtrado a csv
filtro1.to_csv('Filtro1.csv')

from google.colab import files

files.download('Filtro1.csv')
```

```python
#Filtro por filas
filtro2 = df.iloc[ [7,12,66,999], : ] #Muestra Filas no consecutivas 7,12,666,999
filtro2
```

```python
#Convertir archivo filtrado a csv
filtro2.to_csv('Filtro2.csv')

from google.colab import files

files.download('Filtro2.csv')
```

```python
#Filtro por columnas
filtro3 = df.iloc[ : , [0,1,4,5,11] ] #Columnas no consecutivas 0,1,4,5,11
filtro3
```

```python
#Convertir archivo filtrado a csv
filtro3.to_csv('Filtro3.csv')

from google.colab import files

files.download('Filtro3.csv')
```

```python
df1 = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx', index_col=0)
df1
```

```python
#Filtro por cabecera
filtro5 = df.head(15) #Muestra los primeros 15 registros
filtro5
```

```python
#Convertir archivo filtrado a csv
filtro5.to_csv('Filtro5.csv')

from google.colab import files

files.download('Filtro5.csv')
```

```python
#Filtro por comparacion
filtro6 = df[df['COSTO_TOTAL_CALCULADO'] > 10000] #Muestra el monto total que sea mayor de cinco cifras
filtro6
```

```python
#Convertir archivo filtrado a csv
filtro6.to_csv('Filtro6.csv')

from google.colab import files

files.download('Filtro6.csv')
```

```python
#Filtro Y
filtro7 = df[(df['COSTO_TOTAL_CALCULADO'] < 9000) & (df['FECHA_DOC'] == '2022-12-07')] #Muestra los registros de costo total menor a 9000 y en la fecha 2022-12-07
filtro7
```

```python
#Convertir archivo filtrado a csv
filtro7.to_csv('Filtro7.csv')

from google.colab import files

files.download('Filtro7.csv')
```

```python
#Filtro O
filtro8 = df[(df['COSTO_TOTAL_CALCULADO'] < 9000) | (df['FECHA_DOC'] == '2022-12-07')] #Muestra los registros del monto total menor a 9000 o en la fecha 2022-12-07
filtro8
```

```python
#Convertir archivo filtrado a csv
filtro8.to_csv('Filtro8.csv')

from google.colab import files

files.download('Filtro8.csv')
```

```python
#Filtro NOT
filtro9 = df[~(df['COSTO_TOTAL_CALCULADO'] >= 9000) & ~(df['FECHA_DOC'] == '2022-12-07')] #Muestra los registros con un monto total que NO sea mayor o igual a 9000 y que NO se encuentre en la fecha 2022-12-07
filtro9
```

```python
#Convertir archivo filtrado a csv
filtro9.to_csv('Filtro9.csv')

from google.colab import files

files.download('Filtro9.csv')
```
