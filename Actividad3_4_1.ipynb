```python
#Crear DataFrame
df = pd.read_csv('Ventas_totales.csv')
df.head(5)
```

```python
#Conocer las columnas con valores nulos
valores_nulos = df.isnull().sum()
valores_nulos
```

```python
#Se realiza una copia del DataFrame
df1 = df.copy()
```

```python
#Sustituir valores nulos con la media
df1['salon_ventas'] = df1['salon_ventas'].fillna(round(df['salon_ventas'].mean(),2))
df1 #Se sustituyen los valores nulos por el promedio de la columna (del Frame original para evitar errores)
# Se eligio este metodo pues resulta util al establecer un valor concordante con el resto de ventas
```

```python
#Revisar si los valores nulos fueron reemplazados
valores_nulos = df1.isnull().sum()
valores_nulos
```

```python
#Convertir DataFrame a csv
df1.to_csv('Sustitucion1.csv')

#Descargar archivo
from google.colab import files
files.download('Sustitucion1.csv')
```

```python
#Sustituir valores nulos con valores NO nulos hacia adelante ('forward fill')
df1['tarjetas_debito'] = df1['tarjetas_debito'].fillna(method='ffill')
df1 #Se utiliza el metodo forward fill para sustituir el unico valor nulo por el que tiene delante
```

```python
#Revisar si los valores nulos fueron reemplazados
valores_nulos = df1.isnull().sum()
valores_nulos
```
